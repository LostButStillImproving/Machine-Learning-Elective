```python
import pandas as pd

startups = pd.read_csv("50_Startups.csv")
startups = pd.DataFrame(startups)
startups = startups.drop("State", axis = 1)
startups.head()
```

```python
startups.describe()
```

```python
startups.plot(kind="scatter", x = "Marketing Spend", y = "Profit")
```

```python
startups.plot(kind="scatter", y = "Profit", x = "R&D Spend")
```

```python
corr_matrix = startups.corr()
```

```python
corr_matrix["Profit"].sort_values(ascending=False)
```

```python
from pandas.plotting import scatter_matrix

attributes = ["Profit", "R&D Spend", "Marketing Spend",
              "Administration"]
scatter_matrix(startups[attributes], figsize=(12, 7))
```
